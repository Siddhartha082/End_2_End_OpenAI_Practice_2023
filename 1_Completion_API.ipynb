#  temperature parameter 
is the setting in OpenAI that affects the randomness of the output. A lower temperature will result in more predictable output, while a higher temperature will result in more random output.

```python
response = openai.Completion.create(
    model= 'text-curie-001',
    prompt = 'Explain machine learning to a 5 years old kid?',
    max_tokens = 30,
    temperature  = 1,
    n=2
)

response
```

```python
response = openai.Completion.create(
    model= 'text-curie-001',
    prompt = 'Explain machine learning to a 5 years old kid?',
    max_tokens = 30,
    temperature  = 0.6,
    n=1
)

response
```

```python
print(response['choices'][0]['text'])
```
